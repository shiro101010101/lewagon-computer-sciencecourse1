```python
import string

def punc_lower(data):
    data=str(data)
    for punctuation in string.punctuation:
        data=data.replace(punctuation," ")
    return data

df["clean_review_comment_message"]=df["review_comment_message"].apply(punc_lower)
    
```

```python
df["clean_review_comment_message"]=df["clean_review_comment_message"].dropna(axis=0)
```

```python
df
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB

pipeline=Pipeline([
    ("tfidf",TfidfVectorizer()),
    ("nb",MultinomialNB())
])

parameters = {
    'tfidf__ngram_range': ((1,1), (2,2)),
    'nb__alpha': (0.1,1),}


grid_search = GridSearchCV(pipeline, parameters, n_jobs=-1, 
                           verbose=1, scoring = "accuracy", 
                           refit=True, cv=5)

grid_search.fit(df["clean_review_comment_message"],df.review_score)
```

```python
i = [-2, -1, 0]
print(map(abs, i))
# <map object at 0x10651a400>

print(type(map(abs, i)))
# <class 'map'>
```

```python
for i in map(abs, i):
    print(i)
# 2
# 1
# 0
```

```python
l = [-2, -1, 0]
print(list(map(lambda x: x**2, l)))
# [4, 1, 0]
```

```python
def square(x):
    return x**2

print(list(map(square, l)))
# [4, 1, 0]
```

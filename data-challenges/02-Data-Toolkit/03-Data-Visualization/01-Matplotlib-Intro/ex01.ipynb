## Plotting a curved line

Plot the 3 different curves on the same graph, still with pyplot.plot()

```python
x1 = [-10,0,10]
y1 = [5,1,5]

x2 = [-10,-5,0,5,10]
y2 = [5,2,1,2,5]

x3 = [-10,-7.5,-5,-2.5,0,2.5,5,7.5,10]
y3 = [5,3,2,1.2,1,1.2,2,3,5]

# YOUR PLOT HERE
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.show()
```
